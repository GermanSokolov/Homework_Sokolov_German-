```python
import csv
```

```python
def teachers_marks (teachers_id, group_id):
    students_id_list=[]
    subjects_dict={}
    all_group_results=[]
    with open('students.csv', 'r', encoding='1251') as students_file:
        students_reader=csv.reader(students_file)
        for students_row in students_reader:
            useless_list=students_row[0].split(';')
            if (int(useless_list[-1])==group_id):
                students_id_list.append(int(usless_list[0])
    with open('results.csv', 'r', encoding='1251') as results_file:
        results_reader=csv.reader(results_file)
        for result_row in results_reader:
            useless_result=result_row[0].split(';')
            if ((int(useless_result[2])) in students_id_list) and ((int(useless_result[2])==teachers_id):
                all_group_results.append(int(useless_result[-1]))
    with open('subjects.csv', 'r', encoding='1251') as subjects_file:
        sucjects_reader=csv.reader(subjects_file)
        for subject_row in subject_reader:
            useless_subject=subject_row[0].split(';')
            subjects_dict[useless_subject[0]]=useless_subject[-1]
    teacher_students_results={}
    for i in all_group_results:
        b=0
        if i>85:
            m=5
        if 70<i<86:
            m=4
        if 60<i<71:
            m=3
        if i<66:
            m=2
        for k in all_group_results:
            if k>85:
                n=5
            if 70<k<86:
                n=4
            if 60<k<71:
                n=3
            if k<66:
                n=2
            if m==n:
                b+=1
                all_group_results.remove(k)
        teacher_students_results[i]=b
    return teacher_students_results
```
